```python
import os
import cv2
import numpy as np
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from skimage.feature import greycomatrix, greycoprops
from skimage import exposure

# Define the path to the directory where the images are stored
path = r'C:\Users\anves\Downloads\hj3vvx5946-1\hj3vvx5946-1\Splited_dataset\train\Toyota_Vitz_2010'

# Get a list of all the files in the directory
files = os.listdir(path)

# Load the pre-trained ResNet50 model
model = ResNet50(weights='imagenet')

# Initialize lists to store the extracted features and image file names
features = []
filenames = []

# Loop over the list of files and load each image
for filename in files:
    # Load the image using the OpenCV library
    img = cv2.imread(os.path.join(path, filename))
    
    # Resize the image to 224x224 pixels
    img = cv2.resize(img, (224, 224))
    
    # Preprocess the image
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    
    # Use the model to predict the type and color of the car
    preds = model.predict(x)
    preds = decode_predictions(preds, top=1)[0]
    car_type = preds[0][1]
    color = preds[0][2]
    
    # Extract edge features using Canny edge detection
    edges = cv2.Canny(img, 100, 200)
    edge_density = np.sum(edges) / (img.shape[0] * img.shape[1])
    
    # Extract color histogram features using HSV color space
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([hsv], [0, 1], None, [10, 10], [0, 180, 0, 256])
    hist = cv2.normalize(hist, hist).flatten()
    
    # Extract texture features using Gray-Level Co-occurrence Matrix (GLCM)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    glcm = greycomatrix(gray, [5], [0], levels=256, symmetric=True, normed=True)
    contrast = greycoprops(glcm, 'contrast')[0, 0]
    energy = greycoprops(glcm, 'energy')[0, 0]
    
    # Add the extracted features to the features list
    features.append([car_type, color, edge_density, hist, contrast, energy])
    
    # Store the image file name
    filenames.append(filename)

# Print the number of images loaded and features extracted
print("Number of images loaded:", len(filenames))
print("Number of features extracted:", len(features[0]))
```
