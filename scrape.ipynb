```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

name = []
Kilometers_Driven = []
dealer_name = []
price = []
transmission = []
#New Delhi, Mumbai, Bangalore, Hyderabad, Ahmedabad, Pune, Jaipur, Kolkata
website_pages = ['https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=2&pinId=110001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=2378&pinId=400001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=4709&pinId=560001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=3686&pinId=500001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=1692&pinId=380001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=2423&pinId=411001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=2130&pinId=302001','https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=777&pinId=700001']


for url in website_pages:
    url_template = url
    # Loop through all the pages of results
    for page in range(1, 100):
            url = url_template.format(page)
            response = requests.get(url)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Find all the car cards on the page
            car_cards = soup.find_all('div',{'class': '_2z-Yu'})
            
            # Loop through each car card and extract the information you need
            for car_card in car_cards:
                #name
                try:
                    name.append(car_card.find('h2').get_text())
                except:
                    name.append('n/a')
                
                #price
                try:
                    price.append(car_card.find('div',{'class':'_18ToE'}).get_text())
                except:
                    price.append('n/a')

                #Kilometers_Driven
                try:
                    Kilometers_Driven.append(car_card.find('ul',{'class':'_13yb6'}).get_text())
                except:
                    Kilometers_Driven.append('n/a')
                
                #transmission
                try:
                    transmission.append(car_card.find('ul',{'class':'_1hOnS'}).get_text())
                except:
                    transmission.append('n/a')


```

```python
actual_name = []
for i in name:
    car_name = i[5:]
    actual_name.append(car_name)


len(actual_name)
```

```python
string = "2019 Hyundai Elite i20"
car_name = string[5:]
print(car_name)
```

```python
#extracting year
years = []
for car in name:
    match = re.search(r'\b\d{4}\b', car)
    if match:
        years.append(match.group())

len(years)
```

```python
Kilometers_Driven = [s.replace('km', '') for s in Kilometers_Driven]
print(Kilometers_Driven)
```

```python
# extracting actual km driven
actual_km_driven = []

for string in Kilometers_Driven:
    km = string.split(' ')[0]
    actual_km_driven.append(km.replace(',', ''))

print(actual_km_driven)
```

```python
#extracting owner type
owner_type = []
for data in Kilometers_Driven:
    match = re.search(r'\d+(st|nd|rd|th)\s+Owner', data)
    if match:
        owner_type.append(match.group())
    else:
        owner_type.append('Unknown')


owner_types = [ot.replace("1st Owner", "First").replace("2nd Owner", "Second").replace("3rd Owner", "Third") for ot in owner_type]
print(owner_types)
```

```python

fuel_type = []

for fuel in Kilometers_Driven:
    if "Petrol" in fuel:
        fuelt = "Petrol"
        fuel_type.append(fuelt)
    elif "Diesel" in fuel:
        fuelt = "Diesel"
        fuel_type.append(fuelt)
    elif "LPG" in fuel:
        fuelt = "LPG"
        fuel_type.append(fuelt)
    elif "CNG" in fuel:
        fuelt = "CNG"
        fuel_type.append(fuelt)

print(fuel_type)
```

```python
TorM = []
for i in transmission:
    if "Manual" in i:
        t = "Manual"
        TorM.append(t)
    elif "Automatic" in i:
        t = "Automatic"
        TorM.append(t)

print(TorM)
```

```python
actual_price = []
for i in price:
    value = int(i.split('₹')[1].replace(',', ''))
    actual_price.append(value)

print(actual_price)
```

```python
# Create a pandas DataFrame from the extracted data
df = pd.DataFrame({'Name': actual_name,'Year': years,'Kilometers_Driven':actual_km_driven,'Fuel_Type':fuel_type,'Transmission':TorM,'Owner_Type': owner_types ,'Price': actual_price})
# df['Price'] = df['Price'].apply(lambda x:x.strip('₹'))
# Print the DataFrame
print(df)
```

```python
df.to_csv('Scraped_data.csv', index=False)
```
